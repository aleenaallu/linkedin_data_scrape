```python
import os ,time
from selenium import webdriver
from selenium.webdriver.common.by import By

absolute_path = os.getcwd()
relative_path_chrome = "chrome-win/chrome.exe"
relative_path_driver = "chrome-win/chromedriver.exe"
chrome_location = os.path.join(absolute_path, relative_path_chrome)
chromedriver_location = os.path.join(absolute_path, relative_path_driver)
# Set up Chrome options
options = webdriver.ChromeOptions()
options.binary_location = chrome_location
chrome_driver_binary = chromedriver_location

# Initialize Chrome webdriver with options
driver = webdriver.Chrome(options=options)

# Open LinkedIn job search page
link ="https://www.linkedin.com/jobs/search/?currentJobId=3894729962&keywords=Marketing"
# link = "https://www.linkedin.com/jobs/search/?currentJobId=3863211427&geoId=100811329&keywords=marketing%20data%20analyst&location=Karnataka%2C%20India&origin=JOB_SEARCH_PAGE_OTHER_ENTRY&refresh=true&spellCorrectionEnabled=false"
driver.get(link)

# Find the <ul> element containing the job listings using the given XPath
job_listings = driver.find_element(By.XPATH, '//*[@id="main-content"]/section[2]/ul')

# Find all <li> elements within the <ul> element
job_elements = job_listings.find_elements(By.TAG_NAME, 'li')

# Iterate over each <li> element to extract the job title
for job_element in job_elements:
    # Find the <span> element with the class 'sr-only' to extract the job title
    job_title_element = job_element.find_element(By.CLASS_NAME, 'sr-only')
    job_title = job_title_element.text

    # Find the <div> element with the class 'base-search-card_metadata' to extract the address
    # address_element = job_element.find_element(By.CLASS_NAME, 'base-search-card_metadata')
    # address = address_element.text
    
    
    # Find the <a> element with the class 'base-card__full-link' to extract the job link
    link_element = job_element.find_element(By.CLASS_NAME, 'base-card__full-link')
    job_link = link_element.get_attribute('href')


    print("Job Title:", job_title)
    # print("Address:", address)
    print("job url:",job_link)
    print()  # Add an empty line for better readability

    
# Close the browser

driver.quit()
```

```python
import csv
import os ,time
from selenium import webdriver
from selenium.webdriver.common.by import By

absolute_path = os.getcwd()
relative_path_chrome = "chrome-win/chrome.exe"
relative_path_driver = "chrome-win/chromedriver.exe"
chrome_location = os.path.join(absolute_path, relative_path_chrome)
chromedriver_location = os.path.join(absolute_path, relative_path_driver)
# Set up Chrome options
options = webdriver.ChromeOptions()
options.binary_location = chrome_location
chrome_driver_binary = chromedriver_location

# Initialize Chrome webdriver with options
driver = webdriver.Chrome(options=options)

# Open LinkedIn job search page
link ="https://www.linkedin.com/jobs/search/?currentJobId=3894729962&keywords=Marketing"
# link = "https://www.linkedin.com/jobs/search/?currentJobId=3863211427&geoId=100811329&keywords=marketing%20data%20analyst&location=Karnataka%2C%20India&origin=JOB_SEARCH_PAGE_OTHER_ENTRY&refresh=true&spellCorrectionEnabled=false"
driver.get(link)

# Find the <ul> element containing the job listings using the given XPath
job_listings = driver.find_element(By.XPATH, '//*[@id="main-content"]/section[2]/ul')

# Find all <li> elements within the <ul> element
job_elements = job_listings.find_elements(By.TAG_NAME, 'li')

# Open a CSV file in write mode
with open('jobs.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(['Job Title', 'Job Link'])
    
    # Iterate over each <li> element to extract the job title, address, and link
    for job_element in job_elements:
        # Find the <span> element with the class 'sr-only' to extract the job title
        job_title_element = job_element.find_element(By.CLASS_NAME, 'sr-only')
        job_title = job_title_element.text
        
        # Find the <div> element with the class 'base-search-card_metadata' to extract the address
        # address_element = job_element.find_element(By.CLASS_NAME, 'base-search-card_metadata')
        # address = address_element.text
        
        # Find the <a> element with the class 'base-card__full-link' to extract the job link
        link_element = job_element.find_element(By.CLASS_NAME, 'base-card__full-link')
        job_link = link_element.get_attribute('href')
        
        # Write the job data to the CSV file
        writer.writerow([job_title, job_link])
```
